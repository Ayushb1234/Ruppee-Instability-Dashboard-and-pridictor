```python
import pandas as pd
import glob

files = glob.glob(
    r"C:\Users\ka843\Coding\Ruppee-Instability-Dashboard-and-pridictor\data\raw\fpi\*.xls"
)

dfs = []

for file in files:

    # read fake excel html table
    temp_df = pd.read_html(file)[0]

    # flatten multilevel columns
    if isinstance(temp_df.columns, pd.MultiIndex):

        temp_df.columns = [
            '_'.join(
                [str(i) for i in col if str(i) != 'nan']
            ).strip()

            for col in temp_df.columns
        ]

    else:

        temp_df.columns = (
            temp_df.columns
            .astype(str)
            .str.strip()
        )

    dfs.append(temp_df)

# merge all files
fpi = pd.concat(
    dfs,
    ignore_index=True
)

print(fpi.head())
print(fpi.columns)
print(fpi.shape)
```

```python
print(fpi.head())
print(fpi.columns)
print(fpi.shape)
```

```python
fpi.columns = [
    'Date',
    'EquityFlow',
    'DebtFlow',
    'NetFlow'
]
```

```python
fpi['Date'] = pd.to_datetime(
    fpi['Date'],
    errors='coerce'
)
```

```python
numeric_cols = [
    'EquityFlow',
    'DebtFlow',
    'NetFlow'
]

for col in numeric_cols:

    fpi[col] = (
        fpi[col]
        .astype(str)
        .str.replace(',', '', regex=False)
        .str.strip()
    )

    fpi[col] = pd.to_numeric(
        fpi[col],
        errors='coerce'
    )
```

```python
fpi.drop_duplicates(
    subset='Date',
    inplace=True
)
```

```python
fpi.sort_values(
    'Date',
    inplace=True
)
```

```python
fpi.reset_index(
    drop=True,
    inplace=True
)
```

```python
fpi['fii_outflow'] = (
    fpi['NetFlow'] < 0
).astype(int)
```

```python
fpi['flow_change'] = (
    fpi['NetFlow']
    .diff()
)
```

```python
fpi['flow_ma_7'] = (
    fpi['NetFlow']
    .rolling(7)
    .mean()
)
```

```python
fpi['panic_outflow'] = (
    fpi['NetFlow'] < -5000
).astype(int)
```

```python
print(fpi.head())
print(fpi.info())
print(fpi.isnull().sum())
```

```python
fpi.to_csv(
    "data/processed/fpi_master.csv",
    index=False
)
```
