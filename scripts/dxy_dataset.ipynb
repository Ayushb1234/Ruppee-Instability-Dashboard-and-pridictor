```python
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
dxy = yf.download(
    "DX-Y.NYB",
    start="2015-01-01",
    auto_adjust=False
)

dxy.reset_index(inplace=True)
```

```python
dxy.columns = dxy.columns.get_level_values(0)

dxy.rename(columns={
    'Close':'DXY'
}, inplace=True)

dxy['Date'] = pd.to_datetime(
    dxy['Date']
)

numeric_cols = [
    'Open',
    'High',
    'Low',
    'DXY',
    'Volume'
]

for col in numeric_cols:
    dxy[col] = pd.to_numeric(
        dxy[col],
        errors='coerce'
    )

dxy.ffill(inplace=True)
```

```python
dxy['dxy_change'] = (
    dxy['DXY']
    .pct_change()
)
```
