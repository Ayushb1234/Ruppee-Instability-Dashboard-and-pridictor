```python
import pandas as pd
```

```python
usd = pd.read_csv(
    "C:\\Users\\ka843\\Coding\\Ruppee-Instability-Dashboard-and-pridictor\\data\\processed\\usd_inr.csv"
)

oil = pd.read_csv(
    "C:\\Users\\ka843\\Coding\\Ruppee-Instability-Dashboard-and-pridictor\\data\\processed\\oil_data.csv"
)

dxy = pd.read_csv(
    "C:\\Users\\ka843\\Coding\\Ruppee-Instability-Dashboard-and-pridictor\\data\\processed\\dxy_data.csv"
)

nifty = pd.read_csv(
    "C:\\Users\\ka843\\Coding\\Ruppee-Instability-Dashboard-and-pridictor\\data\\processed\\nifty_data.csv"
)

forex = pd.read_csv(
    "C:\\Users\\ka843\\Coding\\Ruppee-Instability-Dashboard-and-pridictor\\data\\processed\\forex_reserves.csv"
)

inflation = pd.read_csv(
    "C:\\Users\\ka843\\Coding\\Ruppee-Instability-Dashboard-and-pridictor\\data\\processed\\inflation.csv"
)

fpi = pd.read_csv(
    "C:\\Users\\ka843\\Coding\\Ruppee-Instability-Dashboard-and-pridictor\\data\\processed\\fpi_master.csv"
)
```

```python
datasets = [
    usd,
    oil,
    dxy,
    nifty,
    forex,
    inflation,
    fpi
]

for df in datasets:

    df['Date'] = pd.to_datetime(
        df['Date']
    )
```

```python
usd = usd[['Date','USD_INR']]

oil = oil[['Date','Oil']]

dxy = dxy[['Date','DXY']]

nifty = nifty[['Date','NIFTY']]

forex = forex[[
    'Date',
    'ForexReserves',
    'GoldReserves',
    'TotalReserves'
]]

inflation = inflation[[
    'Date',
    'Inflation'
]]

fpi = fpi[[
    'Date',
    'NetFlow'
]]
```

```python
master = usd.copy()
```

```python
master = master.merge(
    oil,
    on='Date',
    how='left'
)

master = master.merge(
    dxy,
    on='Date',
    how='left'
)

master = master.merge(
    nifty,
    on='Date',
    how='left'
)

master = master.merge(
    forex,
    on='Date',
    how='left'
)

master = master.merge(
    inflation,
    on='Date',
    how='left'
)

master = master.merge(
    fpi,
    on='Date',
    how='left'
)
```

```python
master.sort_values(
    'Date',
    inplace=True
)
```

```python
master.ffill(
    inplace=True
)
```

```python
master.dropna(
    inplace=True
)
```

```python
master['INR_Return'] = (
    master['USD_INR']
    .pct_change()
)
```

```python
master['Oil_Change'] = (
    master['Oil']
    .pct_change()
)
```

```python
master['DXY_Change'] = (
    master['DXY']
    .pct_change()
)
```

```python
master['INR_Volatility_7D'] = (
    master['USD_INR']
    .rolling(7)
    .std()
)
```

```python
master['CrisisFlag'] = (
    (
        master['INR_Return'] > 0.01
    )
    &
    (
        master['Oil_Change'] > 0.03
    )
).astype(int)
```

```python
master.dropna(
    inplace=True
)

master.reset_index(
    drop=True,
    inplace=True
)
```
