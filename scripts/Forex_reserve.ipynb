```python
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
nifty = yf.download(
    "^NSEI",
    start="2015-01-01",
    auto_adjust=False
)

nifty.reset_index(inplace=True)
```

```python
nifty.columns = nifty.columns.get_level_values(0)

nifty.rename(columns={
    'Close':'NIFTY'
}, inplace=True)

nifty['Date'] = pd.to_datetime(
    nifty['Date']
)

numeric_cols = [
    'Open',
    'High',
    'Low',
    'NIFTY',
    'Volume'
]

for col in numeric_cols:
    nifty[col] = pd.to_numeric(
        nifty[col],
        errors='coerce'
    )

nifty.ffill(inplace=True)
```

```python
nifty['market_return'] = (
    nifty['NIFTY']
    .pct_change()
)
```
