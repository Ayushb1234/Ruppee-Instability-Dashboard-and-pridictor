```python
forex = pd.read_csv(
    "data/raw/forex_reserves.csv"
)
```

```python
forex['Date'] = pd.to_datetime(
    forex['Date']
)

cols = [
    'ForexReserves',
    'GoldReserves'
]

for col in cols:

    forex[col] = (
        forex[col]
        .astype(str)
        .str.replace(',', '')
        .str.replace('B', '')
    )

    forex[col] = pd.to_numeric(
        forex[col],
        errors='coerce'
    )

forex.fillna(method='ffill', inplace=True)
```

```python
forex.set_index('Date', inplace=True)

forex = forex.resample(
    'D'
).ffill()

forex.reset_index(inplace=True)
```

```python
forex.to_csv(
    "data/processed/forex.csv",
    index=False
)
```

```python
import pandas as pd

# read excel
df = pd.read_excel(
    r"C:\Users\ka843\Coding\Ruppee-Instability-Dashboard-and-pridictor\data\raw\Foreign Exchange Reserves.xlsx",
    skiprows=4,
    header=None
)

# keep important columns
df = df[[0,1,3,5,11]]

# rename columns
df.columns = [
    'Year',
    'Month',
    'ForexReserves',
    'GoldReserves',
    'TotalReserves'
]

# remove empty rows
df.dropna(
    subset=['Year'],
    inplace=True
)

# month mapping
month_map = {
    'JAN':'01',
    'FEB':'02',
    'MAR':'03',
    'APR':'04',
    'MAY':'05',
    'JUN':'06',
    'JUL':'07',
    'AUG':'08',
    'SEP':'09',
    'OCT':'10',
    'NOV':'11',
    'DEC':'12'
}

# create date
df['Month'] = df['Month'].map(month_map)

df['Date'] = pd.to_datetime(
    df['Year'].astype(str)
    + '-'
    + df['Month']
    + '-01'
)

# clean numeric columns
numeric_cols = [
    'ForexReserves',
    'GoldReserves',
    'TotalReserves'
]

for col in numeric_cols:

    df[col] = (
        df[col]
        .astype(str)
        .str.replace(',', '')
    )

    df[col] = pd.to_numeric(
        df[col],
        errors='coerce'
    )

# keep final columns
df = df[[
    'Date',
    'ForexReserves',
    'GoldReserves',
    'TotalReserves'
]]

# sort
df.sort_values(
    'Date',
    inplace=True
)

# reset index
df.reset_index(
    drop=True,
    inplace=True
)

# save clean csv
df.to_csv(
    r"C:\Users\ka843\Coding\Ruppee-Instability-Dashboard-and-pridictor\data\processed\forex_reserves.csv",
    index=False
)

print(df.head())
print(df.info())
```

```python
print(df.head())
```

```python
df['Month'] = (
    df['Month']
    .astype(str)
    .str.strip()
    .str.upper()
)
```

```python
month_map = {
    'JAN':'01',
    'FEB':'02',
    'MAR':'03',
    'APR':'04',
    'MAY':'05',
    'JUN':'06',
    'JUL':'07',
    'AUG':'08',
    'SEP':'09',
    'OCT':'10',
    'NOV':'11',
    'DEC':'12'
}
```

```python
df['MonthNum'] = df['Month'].map(month_map)

df['Date'] = pd.to_datetime(
    df['Year'].astype(str)
    + '-'
    + df['MonthNum']
    + '-01'
)
```

```python
numeric_cols = [
    'ForexReserves',
    'GoldReserves',
    'TotalReserves'
]

for col in numeric_cols:

    df[col] = (
        df[col]
        .astype(str)
        .str.replace(',', '', regex=False)
        .str.strip()
    )

    df[col] = pd.to_numeric(
        df[col],
        errors='coerce'
    )
```

```python
df.sort_values(
    'Date',
    inplace=True
)
```

```python
df.reset_index(
    drop=True,
    inplace=True
)
```
