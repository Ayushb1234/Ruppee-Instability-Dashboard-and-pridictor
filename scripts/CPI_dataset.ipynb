```python
import pandas as pd

df = pd.read_excel(
    r"C:\Users\ka843\Coding\Ruppee-Instability-Dashboard-and-pridictor\data\raw\cpi_1133.xlsx"
)
```

```python
df = df[
    df['state'] == 'All India'
]
```

```python
df = df[[
    'year',
    'month',
    'index',
    'inflation'
]]
```

```python
df.columns = [
    'Year',
    'Month',
    'CPI_Index',
    'Inflation'
]
```

```python
month_map = {
    'January':'01',
    'February':'02',
    'March':'03',
    'April':'04',
    'May':'05',
    'June':'06',
    'July':'07',
    'August':'08',
    'September':'09',
    'October':'10',
    'November':'11',
    'December':'12'
}
```

```python
df['MonthNum'] = (
    df['Month']
    .map(month_map)
)

df['Date'] = pd.to_datetime(
    df['Year'].astype(str)
    + '-'
    + df['MonthNum']
    + '-01'
)
```

```python
numeric_cols = [
    'CPI_Index',
    'Inflation'
]

for col in numeric_cols:

    df[col] = pd.to_numeric(
        df[col],
        errors='coerce'
    )
```

```python
df = df[[
    'Date',
    'CPI_Index',
    'Inflation'
]]
```

```python
df.sort_values(
    'Date',
    inplace=True
)
```

```python
df.reset_index(
    drop=True,
    inplace=True
)
```
