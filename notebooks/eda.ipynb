# Corelation Matrix

```python
corr = master.corr(
    numeric_only=True
)

print(corr)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))

sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm'
)

plt.title("Macroeconomic Correlation Heatmap")

plt.show()
```

```python
plt.figure(figsize=(15,6))

plt.plot(
    master['Date'],
    master['USD_INR']
)

plt.title("USD/INR Trend")
plt.xlabel("Date")
plt.ylabel("USD/INR")

plt.show()
```

```python
plt.figure(figsize=(15,6))

plt.plot(
    master['Date'],
    master['Oil'],
    label='Oil'
)

plt.plot(
    master['Date'],
    master['USD_INR'],
    label='USD/INR'
)

plt.legend()

plt.show()
```
